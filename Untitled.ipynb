```python
from difflib import SequenceMatcher
city=''
def match_city(str,lst):
    str.lower()
    match_city=0.0
    for i in lst:
        matcher = SequenceMatcher(None, str, i)
        match_ratio = matcher.ratio()
        if match_ratio==1:
            city=i
            break
        elif match_city<match_ratio:
            match_city=match_ratio
            city=i
            
    if(match_city<0.75):
        city=''
    return city
```

```python
import re

find=''
def find_cuisine(str1,lst):
    if str1=='':
        return ''
    flag=-1
    match=0
    str1=str1.lower()
    for i in lst:
        res=re.split(',',i)
        for j in res:
            matcher= SequenceMatcher(None, str1, j)
            match_ratio=matcher.ratio()
            if match_ratio>=0.99:
                flag=0
                find=j
                break
            if match<match_ratio:
                match=match_ratio
                find=j
            if str1 in j:
                find=j
                flag=0
                break
            
        if flag==0:
            break
    
    if match<0.85 and flag!=0:
        find=''
    return find

print(find_cuisine('',Zomato_data.cuisines))
```

```python
def score(x,y):
    if x=='':
        return 0
    elif isinstance(x, float) or isinstance(y, float):
        return 0
    else:
        matcher=SequenceMatcher(None,x,y)
        match_ratio=matcher.ratio()
        return match_ratio

def y(city,locality,l_verbose,cuisines):
    a=[]
    b=[]
    c=[]
    d=[]
    for i in Zomato_data.city:
        a.append(score(city,i))
    for i in Zomato_data.locality:
        b.append(score(locality,i))
    for i in Zomato_data.locality_verbose:
        c.append(score(l_verbose,i))
    for i in Zomato_data.cuisines:
        d.append(score(cuisines,i))
    x=pd.DataFrame({'city':a,'locality':b,'locality_verbose':c,'cuisines':d,'cost':Zomato_data.cost})
    return x
```

```python
from sklearn.neighbors import NearestNeighbors

def recommend_restaurant(city,locality,l_verbose,cuisines):
    
    knn=NearestNeighbors(n_neighbors=6,algorithm='auto')
    x=y(city,locality,l_verbose,cuisines)
    x.drop(x[x['cost']==0].index,inplace=True)
    knn.fit(x[['city','locality','locality_verbose','cuisines','cost']])
    min=x.cost.min()
    indices=knn.kneighbors([[1,1,1,1,min]])
    print(indices)
    nearest_res=Zomato_data.iloc[indices[1][0]][['name','aggregate_rating','locality','locality_verbose','cuisines','cost']]
    
    nearest_res=nearest_res.sort_values(by='aggregate_rating',ascending=False)
    
    print(nearest_res.head(10))
```

```python
pivot_table=pd.pivot_table(order,values=['rating','sales_amount','sales_qty','cost'],
                           index=['user_id','r_id']
                           ,aggfunc={'rating':'mean','sales_amount':'sum','sales_qty':'sum','cost':'mean'})
```

```python

def rest_list(user_id):
    data=pivot_table.loc[user_id]
    lst=[]
    for ind in data.index:
        for j in order.index:
            if ind == order['r_id'][j]:
                lst.append(order['cuisine'][j])
                break;
    
    df=pd.DataFrame({'cost':data.cost.mean(),'rating':data.rating.mean(),'sales_amount':data.sales_amount.mean(),'cuisines':lst})
    return df
    #user_ratings = pivot_table.loc[user_id]
```

```python

def best_res_fit(cuisine):
    
    for j in cuisine:
        sim_df=pd.DataFrame();
        for i in restaurant.index:
            if(restaurant['cuisine'][i].find(j) != -1):
                sim_df=sim_df.append(restaurant.loc[i])
                #sim_df=pd.concat([sim_df,restaurant.loc[i]],ignore_index=True)
    
    return (sim_df)
```

```python
def rec_res_on_past_order(user_id):
    
    list_pivot=rest_list(user_id)
    sim_rest=best_res_fit(list_pivot.cuisines)
    sim_rest.astype({'rating':'float','cost':'float'})
    
    
    avg_cost=list_pivot.cost.mean()
    avg_rating=list_pivot.rating.mean()
    avg_rating_count=list_pivot.sales_amount.max()
    
    
    knn=NearestNeighbors(n_neighbors=3,algorithm='auto')
    
    knn.fit(sim_rest[['rating','cost']])
    indices=knn.kneighbors([[avg_rating,avg_cost]])
    
    nearest_res=sim_rest.iloc[indices[1][0]][['r_id', 'name', 'city', 'rating', 'rating_count', 'cost', 'cuisine']]
    
    nearest_res=nearest_res.sort_values(by='rating',ascending=False)
    
    print(nearest_res.head(10))
```
